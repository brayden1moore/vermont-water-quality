## **Now that the tool is ready to go, we can use it.**
Here's how you can find the water quality of a **body of water**:

```
vwq.estimate('Lake Dunmore')
```
Here's how to use it to check near a specific **address**:

```
vwq.estimate('206 Omya W, Florence, VT 05744')
```
Or a well known **place**:

```
vwq.estimate('University of Vermont, Burlington')
```
And if you're curious about exact **coordinates**, you can enter those too. Like this. Note the extra parentheses:
```
vwq.estimate((44.3,-72.5))
```
